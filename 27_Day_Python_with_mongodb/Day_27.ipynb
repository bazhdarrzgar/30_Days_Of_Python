### Query with modifiers

````python
from flask import Flask, render_template
import os
import pymongo

try:
    
    MONGODB_URI = 'mongodb+srv://soyansoon:dUuiEuBaARLwocjH@mycluster.yix1qqc.mongodb.net/?retryWrites=true&w=majority'
    client = pymongo.MongoClient(MONGODB_URI)
    
    
    # accessing the database
    db = client['thirty_days_of_python']
    
    # creating condition that i use just for calling this row that contain this key with this value 
    
    # "$gt": 30   means greater than 30
    
    # this dictionary means the key is   "age"   and the value should be greater than 30 for doing that we use specific dictionary as value like this {"$gt": 30} 
    # this because we want use keyword   "$gt"  as key that means greater than
    # in the end you get dictionary like this   {"age": {"$gt": 30}}
    query = {"age": {"$gt": 30}}
    
    # find this row of data from collesion student of my database   thirty_days_of_python   that match my condition   (query)  variable
    students = db.students.find(query)
    
    # calling all this row that match this row that we get from   student   variable
    for student in students:
        print(student)

        
    app = Flask(__name__)
    
    if __name__ == '__main__':
        port = int(os.environ.get("PORT", 82))
        app.run(debug=True, host='0.0.0.0', port=port)
        
except:
    print('permission denied')
````

# Find with sort

````python
from flask import Flask, render_template
import os
import pymongo

try:
    
    MONGODB_URI = 'mongodb+srv://soyansoon:dUuiEuBaARLwocjH@mycluster.yix1qqc.mongodb.net/?retryWrites=true&w=majority'

    client = pymongo.MongoClient(MONGODB_URI)
    
    
    # accessing the database
    db = client['thirty_days_of_python']     
    
    # sorting by name using   sort()   function by default the sort will in Ascending order
    students = db.students.find().sort('name')
    for student in students:
        print(student)


    # sorting by name in reverse way using   sort()   function with parameter ( -1 ) that indicate for reverse
    students = db.students.find().sort('name',-1)
    for student in students:
        print(student)

    # sorting by age
    students = db.students.find().sort('age')
    for student in students:
        print(student)

    # sorting by age in reverse way using   sort()   function with parameter ( -1 ) that indicate for reverse
    students = db.students.find().sort('age',-1)
    for student in students:
        print(student)

        
    app = Flask(__name__)
    
    if __name__ == '__main__':
        port = int(os.environ.get("PORT", 82))
        app.run(debug=True, host='0.0.0.0', port=port)
        
except:
    print('permision denied')
    
'''

# Ascending order

{'_id': ObjectId('5df68a21f106fe2d315bbc8b'), 'name': 'Asabeneh', 'country': 'Finland', 'city': 'Helsinki', 'age': 250}
{'_id': ObjectId('5df68a23f106fe2d315bbc8c'), 'name': 'David', 'country': 'UK', 'city': 'London', 'age': 34}
{'_id': ObjectId('5df68a23f106fe2d315bbc8d'), 'name': 'John', 'country': 'Sweden', 'city': 'Stockholm', 'age': 28}
{'_id': ObjectId('5df68a23f106fe2d315bbc8e'), 'name': 'Sami', 'country': 'Finland', 'city': 'Helsinki', 'age': 25}


# Descending order

{'_id': ObjectId('5df68a23f106fe2d315bbc8e'), 'name': 'Sami', 'country': 'Finland', 'city': 'Helsinki', 'age': 25}
{'_id': ObjectId('5df68a23f106fe2d315bbc8d'), 'name': 'John', 'country': 'Sweden', 'city': 'Stockholm', 'age': 28}
{'_id': ObjectId('5df68a23f106fe2d315bbc8c'), 'name': 'David', 'country': 'UK', 'city': 'London', 'age': 34}
{'_id': ObjectId('5df68a21f106fe2d315bbc8b'), 'name': 'Asabeneh', 'country': 'Finland', 'city': 'Helsinki', 'age': 250}

'''
````

### Update with query

````python
from flask import Flask, render_template
import os
import pymongo

try:
    MONGODB_URI = 'mongodb+srv://soyansoon:dUuiEuBaARLwocjH@mycluster.yix1qqc.mongodb.net/?retryWrites=true&w=majority'
    
    client = pymongo.MongoClient(MONGODB_URI)

    
    # accessing the database
    db = client['thirty_days_of_python'] 

    # my data that i will change it
    query = {'age': 250}
    # this is the data that i apply process change using   $set   keyword for this change and i used with   update_one()   function
    new_value = {'$set':{'age': 38}}

    # using   update_one()   function to update value or change vlaue in database mongodb
    # remember this is just do for first matching result menas not do that for all the row if you use in for loop you can do it for all
    db.students.update_one(query, new_value)
    # lets check the result if the age is modified
    for student in db.students.find():
        print(student)


    app = Flask(__name__)

    if __name__ == '__main__':
        port = int(os.environ.get("PORT", 82))
        app.run(debug=True, host='0.0.0.0', port=port)
        
except:
    print('permission denied')
    
'''
{'_id': ObjectId('5df68a21f106fe2d315bbc8b'), 'name': 'Asabeneh', 'country': 'Finland', 'city': 'Helsinki', 'age': 38}
{'_id': ObjectId('5df68a23f106fe2d315bbc8c'), 'name': 'David', 'country': 'UK', 'city': 'London', 'age': 34}
'''
````

### Drop a collection

````python
from flask import Flask, render_template
import os
import pymongo

try:
    MONGODB_URI = 'mongodb+srv://soyansoon:dUuiEuBaARLwocjH@mycluster.yix1qqc.mongodb.net/?retryWrites=true&w=majority'

    client = pymongo.MongoClient(MONGODB_URI)
    
    # accessing the database
    db = client['thirty_days_of_python']
    
    # delete entire database   'thirty_days_of_python'   using   drop()   function that delete the data of database and database itself
    db.students.drop()
    
    # lets check the result if my data of database is deleted
    for student in db.students.find():
        print(student)
    
    # also checking the database
    print(client.list_database_names())

        
except:
    print('permission denied')
````
